## Functions

Thus far we've used a couple of basic mathematical functions, namely **cos** and **sin**, unsurprisingly there are [many more](http://docs.scipy.org/doc/numpy/reference/routines.math.html). We know how to use import to pull in functions from other people's modules but we can also construct our own.

We'll construct an $x^2$ function:

    def our_x2_function(x):
        return x**2

Let's see if it works:

    our_x2_function(10)

The format for defining our own functions is as follows:

Step 1: write a function *signature*
----

A function *signature* tells you how the function will be called. It starts with the keyword *def*, which tells Python that you are *defining* a function. Then comes a space, the name of your function, an open parenthesis, the comma-separated input parameters for your function, a close parenthesis, and a colon. Here's what a function signature looks like for a function that takes no arguments:

    def my_function():

Here's what a function signature looks like for a function that takes one argument called my_integer:

    def my_function(my_integer):

And one for a function that takes two arguments:

    def my_function(my_integer_1, my_integer_2):

Parameters should have names that usefully describe what they are used for in the function.

Step 2: do useful work inside the function
----

Underneath the function signature you do your useful work. Everything inside the function is indented, so Python knows that it is a part of the function.

We can use the variables passed into the function as parameters, just like we can use variables once we define them outside of functions.

    def add(x, y):
        result = x + y

Step 3: return something
----

If you want to be able to assign a variable to the output of a function, the function has to return that output using the return keyword.

    def add(x, y):
        result = x + y
        return result

or, even shorter:

    def add(x, y):
        return x + y

Once you execute a return, you are done with the function -- you don't get to do any more work. That means if you have a function like this:

    def add(x, y):
        return x + y
        z = x**2 + y**2
        
The line z = x\**2 + y\**2 will never execute

Step 4: use the function
---

Once a function is defined we can use it as many times as we want.

    def add(x, y):
        return x + y

    add(1,2)

    add(2,3)

    a=1
    b=2
    c=add(a,b)
    c

## Interact

Armed with our new ability to build functions we can do something a bit snazzy with our plotting. We can make them interactive! Suppose we have a simple $y=x^2 +c_{0}x$ function that we want plot, let's plot 50 points it for x between 0 and 10, with $c_0$ set to 1: 

    c0=1
    x = linspace(0,10,50)
    y = x**2  + c0*x
    plot(x,y, linestyle='', marker='o')
    show()

```python
c0=1
x = linspace(0,10,50)
y = x**2  + c0*x
plot(x,y, linestyle='', marker='o')
show()
```

Now let's suppose we want to see how plot change as the coefficient infront of x changes. We could make several different plots picking different values for this coefficient. That's a reasonable thing to do but we could also make the figure dynamically change as we vary our coefficient.

```python
from IPython.html.widgets import interact

@interact(c0=[-10,10,0.01])
def interact_poly(c0):
    x = linspace(0, 10, 50)
    y = x**2 + c0*x
    plot(x, y, linestyle='', marker='o')
    show()
```

What we've done is to bundle all the plotting commands inside a function that we've called "interact_poly") where the coefficient we're interested in making interactive is the one of the functions arguments.

We then add the special function '@interact' above the line where we define "interact_poly" inside the @interact function we put the parameter we want to vary which is c0 in our case and we tell it we want to look at c0 between -10 and 10 in steps that 0.01 large. It does the rest for us. We can even use this with our polyfit example above to choose our filtering. Take a look at, then execute the following cell. Quite a lot is happening but you should be able to get the gist of it.

```python
@interact(cutoff_temp=[315,400])
def interact_polyfit(cutoff_temp):
    test_data = loadtxt('ideal_gas_data.txt')
    temp = test_data[:,0]
    pressure = test_data[:,1]

    filtered_temp = temp[temp<=cutoff_temp]
    filtered_pressure= pressure[temp<=cutoff_temp]

    array([filtered_temp, filtered_pressure])
    poly_out = polyfit(filtered_temp,filtered_pressure, 1, cov=True)
    
    params=poly_out[0]
    m=params[0]
    c=params[1]
    
    poly_err = poly_out[1]**0.5
    m_err = poly_err[0][0] 
    
    plot(temp,pressure,marker='o', linestyle='')
    plot(temp, m*temp + c, linestyle='--')
    plot([cutoff_temp,cutoff_temp], [100000,145000])
    show()
    print('The uncertainty in the gradient is', m_err)
```

## Non Linear fitting

We will now turn our attention to fitting the spectrum of a so called black body (essentially an object in perfect thermal equilbrium). First we'll load the data:
    
    spec_data = loadtxt('black_body.txt')

```python
spec_data = loadtxt('black_body.txt')
wavelengths=spec_data[:,0]
intensities=spec_data[:,1]
intensity_err=spec_data[:,2]
```

Let's take a look at it, for completeness we'll add the units these things are measured in as labels:

Because the error bars are so small in most cases they're hidden by the markers. We'll fix that by shrinking the marker size.

    show()

```python
figure(figsize=[18,12])
errorbar(wavelengths, intensities, yerr=intensity_err, linestyle='',marker='o', markersize=2 )
show()
```

Ok now we're good to go. The law that describes this emission is Planck's law:

$$I_T(\lambda) =\frac{2 hc^2}{\lambda^5} \frac{1}{e^{h c/(\lambda k_\mathrm{B}T)} - 1}$$

Don't worry too much about the equation, you will learn a little bit about how finding this equation lead to revolutions in our understanding of matter during your introduction to quantum mechanics. For now we'll just take for granted that it tells us how much light is emitted at a particular wavelength and how this varies with temperature. 

Really when we talk about fitting we mean we're varying the parameters an equation depends upon (in this case the temperature) until what the function predicts is as close as possible to the data we have. We might do this because we want to make new predictions or in many cases we are interested in the parameters themselves. We can use the above equation to turn a spectrometer into a thermometer for black-bodies. 

We'll do that here to figure out how hot the object was that emitted the radiation our experiment captured.

First we need to translate Plancks law into a form the computer understands using our newly learned *function* syntax. Planck's law involves some constants so we'll import those then use them in our function:

    from scipy.constants import h,c,k

    def intensity(wavelength, temperature):
        return 2*h*c**2/wavelength**5 * 1/(e**(h*c/(wavelength*k*temperature))-1)

```python
from scipy.constants import h,c,k
    
def intensity_f(wavelength, temperature):
    return 2*h*c**2/wavelength**5 * 1/(np.e**(h*c/(wavelength*k*temperature))-1)
```

Let's make sure our function is working by plotting the variation of intensity with wavelength for a temperature of 5800 Kelvin (about the temperature of the sun):

    test_wavelengths = linspace(1e-10,2e-6,200)
    test_intenties = intensity_f(test_wavelengths, 5800)
    plot(test_wavelengths, test_intensities)
    show()

```python
test_wavelengths = linspace(1e-8,2e-6,200)
test_intensities = intensity_f(test_wavelengths,5800)
plot(test_wavelengths, test_intensities)
show()
```

(This is actually not such an awful model of the spectral output of the sun, the region of greatest intensity is the visible spectrum spanning 400-700 nm)

Ok now we can fit our function to the data. First we need to import the function for non-linear fitting - **curve_fit**

    from scipy.optimize import curve_fit

```python
from scipy.optimize import curve_fit
```

curve_fit works in a similar way to the polyfit function we used before except that it takes as its first argument the function we are trying to fit:

    curve_fit(intensity_f, wavelengths, intensities)

```python
curve_fit(intensity_f, wavelengths, intensities)
```

It gives us back the coefficients we are interested in and like polyfit gives us a matrix of numbers whose diagonal elements if squared give us the uncertainties assoiated with the coefficients. In this case we only have one coefficient so the covariance matrix is only a single number. In this case though we have a problem: we're getting back a coefficient with the value of one and an uncertainty of infinity! If we use this coefficient and take a look at what intensities our function predicts for the wavelengths in our data set we see:

    intensity_f(wavelengths, 1)

```python
intensity_f(wavelengths, 1)
```

These values are insanely small - hundred of orders of magnitude too small! What's gone wrong? Well we didn't specify a guess for what our parameter should be and that meant curve_fit used it's default guess value, which happens to be 1, and that turned out to be a bad guess. 

The down side of doing a non linear fit is that we need to start off with vaguely sensible numbers otherwise our fitting process can go badly (or not so badly) wrong. Usually you want to try and use some kind of physical knowledge to guide your guess, though sometimes you have to just use trial and error. In our case the default guess means our starting point is a prediction for 1 degree Kelvin which is a very low temperature. We will use the *p0* keyword to specify a guess of 150 Kelvin (which is still a low temperature from the perspective of synthetic chemistry but is not verging on absolute zero)

    curve_fit(intensity_f, wavelengths, intensities, p0=[150])

```python
curve_fit(intensity_f, wavelengths, intensities, p0=[150])
```

Ah much better now, we have a predicted coefficient of ~298 Kelvin i.e. room temperature and the curve_fit is reporting an uncertainty of just 0.16. But actually we can do even better. Just as polyfit can can take into account the uncertainties in our data so that more weight is associated with data points that are more certain, curve_fit can do the same.

```python
curve_fit(intensity_f, wavelengths, intensities, p0=[150], sigma=intensity_err)
```

When we used polyfit we passed in the weights associated with each points, where the weight was $\frac{1}{\sigma^2}$, curve_fit is doing exacty the same thing but we give it the sigma values and it works out the weighting for us. Thus we pass the sigma values directly to the function. 

Doing so means our uncertainty drops to just 0.073

    curve_out = curve_fit(intensity_f, wavelengths, intensities, p0=[100])
    curve_params = curve_out[0]
    curve_cov = curve_out[1]
    curve_cov **0.5

```python
curve_out = curve_fit(intensity_f, wavelengths, intensities, p0=[150], sigma=intensity_err)
predicted_temp = curve_out[0]
predicted_temp_cov = curve_out[1]
predicted_temp_cov **0.5
```

Let's use our fitted function to generate some intensities. We could use the wavelengths that are in our data set but we'd like to produce a smoother curve so we'll pick the same range of values but we'll span them with 200 points. rather than the 43 that are in the dataset

    test_wavelengths = linspace(1e-9,5e-5,100)
    predicted_intensities = intensity_f(test_wavelengths, predicted_temp)
    predicted_intensities

```python
test_wavelengths = linspace(1e-9,5e-5,100)
predicted_intensities = intensity_f(test_wavelengths, predicted_temp)
predicted_intensities
```

And plot out predicted_intensities alongside our original data:

    figure(figsize=[18,12])
    plot(wavelengths, intensities linestyle='',marker='o')
    plot(test_wavelengths, predicted_intensitiesm linestyle='--')
    show()

```python
figure(figsize=[18,12])
errorbar(wavelengths, intensities, yerr=intensity_err, linestyle='',marker='o', markersize=2 )
plot(test_wavelengths, predicted_intensities, linestyle='--')
show()
```

For completeness we'll add the units these things are measured in as labels, and shift the units so that our figure has nicer values along the axes.


**mmm maybe labels that are less like lies?**

```python
wavelengths_new_units = wavelengths/1e-6
intensities_new_units = intensities/1e7
intensity_err_new_units = intensity_err/1e7

figure(figsize=[18,12])
errorbar(wavelengths_new_units, intensities_new_units, yerr=intensity_err_new_units, linestyle='',marker='o', markersize=2, label='Experimental data')
plot(test_wavelengths/1e-6, predicted_intensities/1e7, linestyle='--', label='Theoretical Blackbody')
title('Intensity vs. Wavelength @298K', fontsize=18)
xlabel('Wavelength / $\mu{}m$', fontsize=18)
ylabel(r'Intensities / $1\times{}10^7.W⋅sr^{−1}$', fontsize=18)
xlim(0,50.5)
legend()
show()
```

As we expected from the very low uncertainty associated with our predicted temperature the fit is excellent. We conclude that the body emitting the spectrum our data set captured was at ~298 Kelvin - room temperature.
